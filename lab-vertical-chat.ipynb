```python
from openai import OpenAI
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
```

```python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)

def continue_conversation(messages, temperature=0):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=temperature,
    )
    #print(str(response.choices[0].message["content"]))
    return response.choices[0].message.content
```

```python
def add_prompts_conversation(_):
    #Get the value introduced by the user
    prompt = client_prompt.value_input
    client_prompt.value = ''

    #Append to the context the User prompt.
    context.append({'role':'user', 'content':f"{prompt}"})

    #Get the response.
    response = continue_conversation(context)

    #Add the response to the context.
    context.append({'role':'assistant', 'content':f"{response}"})

    #Update the panels to show the conversation.
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600)))

    return pn.Column(*panels)
```

```python
import panel as pn

# Initialize conversation list
conversation = []

# Function to handle the conversation for each prompt
def handle_prompt(prompt):
    context = [{'role':'user', 'content':f"{prompt}"}]
    response = continue_conversation(context)
    context.append({'role':'assistant', 'content':f"{response}"})
    return response

# Function to add prompts and responses to the conversation
def add_prompts_conversation(event):
    user_input = client_prompt.value
    response = handle_prompt(user_input)
    conversation.append(pn.pane.Markdown(f"**User:** {user_input}"))
    conversation.append(pn.pane.Markdown(f"**Assistant:** {response}"))
    client_prompt.value = ""
    interactive_conversation[:] = conversation

# Create widgets
client_prompt = pn.widgets.TextInput(value="", placeholder='Enter text hereâ€¦')
button_conversation = pn.widgets.Button(name="Talk")

# Bind button to function
button_conversation.on_click(add_prompts_conversation)

# Create interactive conversation panel
interactive_conversation = pn.Column(
    *conversation
)

# Create the dashboard
dashboard = pn.Column(
    client_prompt,
    pn.Row(button_conversation),
    interactive_conversation
)

# Display the dashboard
pn.extension()
dashboard.servable()
```
